**2)** This dataset includes fuel economy data from 1999 and 2008 for 38 popular models of car. It comes from the EPA is useful to see how cars have changed over time. 

```python
# Pandas for managing datasets
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns

#NUMPY
import numpy as np

# Read dataset
cars= sns.load_dataset("mpg")
cars.head()
cars.shape
cars.isnull().any()
cars.dtypes
cars.horsepower.unique()
```

```python
cars.mpg.describe()
```

```python
sns.distplot(cars['mpg'])
```

This overview shows that the mean mpg is approximately 23.513 with values ranging from 9 to 46.6
The histogram shows a slight right skew in the data indicating that more cars have an mpg less than the mean. 

```python
cars.weight.describe()
```

```python
sns.distplot(cars['weight'])
```

This overview details the weight. The mean is about 2970 lbs with values ranging between 1613 and 5140. The histogram shows a right skew in the data, once again indicating that more values are less than the mean. Looking at the histogram, one can see a strong spike in the cars that weight approximately 2000 lbs. 

```python
fig = sns.boxplot(x='cylinders', y="mpg", data = cars)
plt.axhline(cars.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
```

This plot shows the mpg in relation to the number of cylinders a car has. The mpg shows a curved trend. As the number of cylinders increases, the mpg goes up, until 5 cylinders where it starts to decrease. Cylinders represent part of the cars engine system. Generally, an engine with more cylinders produces more power, while an engine with fewer cylinders gets better fuel economy. 

```python
fig = sns.boxplot(x='origin', y="mpg", data = cars)
plt.axhline(cars.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
```

```python
fig = sns.boxplot(x='model_year', y="mpg", data = cars)
plt.axhline(cars.mpg.mean(),color='r',linestyle='dashed',linewidth=2)
```

This plot shows box and whisker plots for each year based on mpg. From the years 1970-1979, there is no real, distinguishable pattern. However, in 1980, there is a clear spike in the mpg which continues for the following two years. One possible explanation for this is that year, the fuel economy law of 1975 or Corporate Average Fuel Economy (CAFE) was put in place by congress. This law set the average new vehicle fuel economy, as weighted by sales, that a manufacturer's fleet must achieve. 

```python
plt.scatter('horsepower', 'acceleration', color = 'red', data = cars)
plt.title("Horsepower vs Acceleration")
plt.xlabel("Horsepower")
plt.ylabel("Acceleration")
#plt.show()
```

This scatter plot shows the negative correlation between horsepower and acceleration. The horsepower decreases as the acceleration increases. 

```python
x = cars['weight']
y = cars['acceleration']
plt.scatter(x, y)
plt.xlabel("Weight")
plt.ylabel("Acceleration")
plt.title("Weight vs Acceleration")

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

A similar relationship occurs with acceleration and weight. As the car gets heavier, the acceleration decreases. This intuitively makes sense because heavier cars would require more power and energy to pick up speed. This in turn requires more resources like fuel.
